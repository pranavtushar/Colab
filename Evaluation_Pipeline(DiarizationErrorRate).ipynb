```python
from pyannote.metrics.diarization import DiarizationErrorRate
from pyannote.metrics.segmentation import *
import pandas as pd
import glob,os

diarizationErrorRate = DiarizationErrorRate()
```

```python
def get_GroundAnnotation(gpath):
  g_csv= pd.read_csv(gpath)
  g_start,g_end,g_speaker_id  = g_csv['source_start'].to_list(),g_csv['source_end'].to_list(),g_csv['speaker_id'].to_list()
  # g_start,g_end,g_speaker_id  = g_csv['whisper_start'].to_list(),g_csv['whisper_end'].to_list(),g_csv['speaker_id'].to_list()

  unique_values = []
  for x in g_speaker_id:
      if x not in unique_values:
          unique_values.append(x)
  speaker_labels = {unique_values[i]: 'SPEAKER_{:02d}'.format(i) for i in range(len(unique_values))}
  newg_speaker_id = ['SPEAKER_{:02d}'.format(list(speaker_labels.keys()).index(x)) if x in list(speaker_labels.keys()) else x for x in g_speaker_id]

  g_timeline = Annotation()
  for i in range(len(g_start)):
      s = g_start[i]
      e = g_end[i]
      id = newg_speaker_id[i]
      segment = Segment(s, e)
      g_timeline[segment] = id
  return g_timeline

def get_PredAnnotation(ppath):
  p_csv= pd.read_csv(ppath)
  p_start, p_end, p_speaker_id  = p_csv['whisper_start'].to_list(),p_csv['whisper_end'].to_list(), p_csv['speaker_id'].to_list()
  # p_start, p_end, p_speaker_id  = p_csv['start'].to_list(),p_csv['end'].to_list(), p_csv['speaker'].to_list()

  p_timeline = Annotation()
  for i in range(len(p_start)):
      s = p_start[i]
      e = p_end[i]
      id = p_speaker_id[i]
      # print(id)
      segment = Segment(s, e)
      p_timeline[segment] = id
  return p_timeline
```

```python
ground_df = '/content/drive/MyDrive/DRZ-GroundTruth/*csv'
pred_df = '/content/drive/MyDrive/DRZ-Predicted/*csv'

DER =[]
for gpath in glob.glob(ground_df):
  for ppath in glob.glob(pred_df):
    if gpath.split('/')[-1].split('.')[0] == ppath.split('/')[-1].split('.')[0]:
      reference = get_GroundAnnotation(gpath)
      hypothesis = get_PredAnnotation(ppath)
      der = diarizationErrorRate(reference, hypothesis)
      DER.append({'ID': ppath.split('/')[-1].split('.')[0],
                  'DER': (der*100)
                 })

sav_df = pd.DataFrame(DER)
sav_df.to_csv('DER.csv')
```
