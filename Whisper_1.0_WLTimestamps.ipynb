```python
from whisper import transcribe,load_model
from datetime import timedelta
import whisper
import torch
import os

model_size = "large-v2"
audio_path = "/content/input_audio.wav"
language = "en"
task = "transcribe"
initial_prompt = ""

transcribe_args={
    'task': task, 'language': language, 'patience': None,
    'length_penalty': None, 'suppress_tokens': '-1',
    'initial_prompt': initial_prompt, 'fp16': False,
    'condition_on_previous_text':True,'word_timestamps':True,
    }

if torch.cuda.is_available():
    model_transcribe = whisper.load_model(model_size).cuda().eval()
else:
    model_transcribe = whisper.load_model(model_size)
```

```python
text = whisper.transcribe(model_transcribe,audio_path,**transcribe_args)
```

```python
for i in text:
  print(i)
print("Getting into Segments..")
print(" ")
for i in text['segments']:
  print(i)
```

```python
t=[]
for i in text['segments']:
  first = i['words']
  for i in first:
    t.append(i)
```

```python
import pandas as pd
csv = pd.DataFrame(t)
print(csv)
# csv.to_csv('/content/WLTimestamps.csv')
```
