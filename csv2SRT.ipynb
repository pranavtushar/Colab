```python
from datetime import timedelta
import pandas as pd
import csv
df = pd.read_csv('/content/oFYSxFlmhbc.csv')
df.head(5)
```

```python
def convert_seconds_to_srt_time(seconds):
    time_obj = timedelta(seconds=seconds)
    hours, remainder = divmod(time_obj.seconds, 3600)
    minutes, seconds = divmod(remainder, 60)
    milliseconds = time_obj.microseconds // 1000
    return f"{hours:02d}:{minutes:02d}:{seconds:02d},{milliseconds:03d}"

def convert_csv_to_srt(csv_file_path, srt_file_path):
    with open(csv_file_path, 'r') as csv_file, open(srt_file_path, 'w') as srt_file:
        reader = csv.reader(csv_file)
        count = 1
        for row in reader:
            try:
                start_time_seconds = float(row[1])
                end_time_seconds = float(row[2])
                text = row[3]
                start_time_srt = convert_seconds_to_srt_time(start_time_seconds)
                end_time_srt = convert_seconds_to_srt_time(end_time_seconds)

                srt_file.write(f"{count}\n")
                srt_file.write(f"{start_time_srt} --> {end_time_srt}\n")
                srt_file.write(f"{text}\n\n")
                count += 1

            except:
              continue
                # print('cannot convert into float:',row[1],row[2],row[3])
    print('SRT File (Subtitle) Created')
```

```python
csv_file_path='/content/oFYSxFlmhbc.csv'
srt_file_path = '/content/oFYSxFlmhbc.srt'
convert_csv_to_srt(csv_file_path, srt_file_path)
```
